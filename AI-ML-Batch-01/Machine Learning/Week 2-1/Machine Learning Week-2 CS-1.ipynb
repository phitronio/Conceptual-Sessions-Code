```python
import numpy as np
```

```python
v1 = np.array([5, 3 ,2])
v2 = np.array([3,5,7])

dot_product = np.dot(v1,v2)

print(dot_product)
```

```python
v1 = np.array([5, 3 ,2])
v2 = np.array([3,5,7])

dist = v1 -v2

euclidean_distance = np.linalg.norm(dist)

print(euclidean_distance)
```

```python
mat1 = np.array([[1,2],
                 [3,4]])

mat2 = np.array([[1 , 2 , 3],
                 [4,5,6]])

dot_product = np.dot(mat1,mat2)

print(dot_product)
```
