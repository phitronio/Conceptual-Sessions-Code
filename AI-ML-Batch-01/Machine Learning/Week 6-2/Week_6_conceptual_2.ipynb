```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor , plot_tree
```

```python
data = {
    'Rooms':[1,2,2,3,4,5],
    'Baths':[1,1,2,2,3,4],
    'Price':[100,200,250,350,450,600]

}

df = pd.DataFrame(data)
df
```

```python
df['pred0'] = df['Price'].mean()

df['res_0']= df['Price'] - df['pred0']
df
```

```python
X = df[['Rooms','Baths']]
y = df['res_0']

plt.figure(figsize=(10,6))
tree_1 = DecisionTreeRegressor()
tree_1.fit(X,y)



plot_tree(tree_1,
          feature_names=['Rooms','Baths'])

plt.show()
```

```python
df['res_1'] = tree_1.predict(X)
df
```

```python
learning_rate = 0.1

df['pred1'] = df['pred0']+ learning_rate * df['res_1']

```

```python
df['res_2']= df['Price'] - df['pred1']

```

```python
X = df[['Rooms','Baths']]
y = df['res_2']

plt.figure(figsize=(10,6))
tree_1 = DecisionTreeRegressor()
tree_1.fit(X,y)



plot_tree(tree_1,
          feature_names=['Rooms','Baths'])

plt.show()
```

```python
df['res_3'] = tree_1.predict(X)

learning_rate = 0.1

df['pred2'] = df['pred0'] + learning_rate * df['res_1'] + learning_rate * df['res_3']

df
```
