```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#line plot



days = [1,2,3,4,5]
temperature = [30,32,31,29,35]

plt.plot(days, temperature)

#xticks, y ticks

plt.xticks([1,2,3,4,5])

plt.title("Temperature Trend")

plt.xlabel("Day")
plt.ylabel("Temperature")

plt.show()
```

```python
# line plot on dataframe

data = {
    'Day' : [1,2,3,4,5],
    'Sales' : [200,250,300,280,320]
}

df = pd.DataFrame(data)

df

df.plot( x = 'Day' ,y ='Sales', kind='line')

plt.show()
```

```python
# customizing line plot

plt.plot( days, temperature , color='red', linestyle='--', marker='o')

plt.grid(True)
plt.show()
```

```python
# Scatter plot

hours = [1,2,3,4,5]
marks = [40,50,65,70,85]

plt.scatter(hours,marks)

plt.show()
```

```python
# histogram

marks = [40,50,60,70,80,90,100,60,70,80]

plt.hist(marks,bins=10)

plt.show()
```

```python
# Bar chart

subject = ["Math","CS","Physics"]
students = [50,80,40]

plt.bar(subject,students)

plt.show()
```

```python
# pie chart

lables = ['Rent','Food','Transport']
values = [50,30,20]

plt.pie(values,labels=lables, autopct='%1.1f%%')

plt.show()
```

# Seaborn

```python
import seaborn as sns

tips = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

tips
```

```python
#lineplot

sns.lineplot(x="size", y="total_bill", data = tips, errorbar=None)
```

```python
# Scatterplot

sns.scatterplot(x= "total_bill", y = "tip", hue="sex" ,data=tips)
```

```python
tips['time'].unique()
```

```python
# facet plot

sns.relplot(x="total_bill",y="tip",col="time", data=tips)
```

```python
# Histogram

sns.histplot(tips['total_bill'])
```

```python
# KDE plot

sns.kdeplot(tips['total_bill'])
```

```python
# count plot

sns.countplot(x='sex',data=tips)
```

```python
#bar plot

sns.barplot(x='day',y='total_bill',data=tips)
```

```python
# regplot

sns.regplot(x="total_bill",y="tip",data=tips)
```

```python
# pairplot and jointplot
```

```python
sns.pairplot(tips)
```

```python
#joint plot

sns.jointplot(x="total_bill",y='tip',data=tips)
```
