# Map

map(function, iterable, ...)

Examples of Iterables
- Lists: [1, 2, 3]
- Tuples: (1, 2, 3)
- Strings: "hello"
- Dictionaries: {'a': 1, 'b': 2}
- Sets: {1, 2, 3}
- Files: open('file.txt')

```python
arr = [1,2,3,4,5,6] # [10, 20, 30, 40, 50, 60]

def fun1(x):
  return x*20

arr2=list(map(fun1, arr))
print(arr2)
```

```python
arr = [1,2,3,4,5,6] # [10, 20, 30, 40, 50, 60]


arr2=list(map(lambda x: x*20, arr))
print(arr2)
```

```python
scores = [85, 92, 78, 65, 45]


def grade(x):
  if x >=80 : return f"Score is {x} and A+"
  if x >=70 : return f"Score is {x} and A"
  if x >=60 : return f"Score is {x} and A-"
  else : return f"Score is {x} and F"

ans = list( map(grade, scores) )
print(ans)
```

# Filter
filter(function, iterable)

```python
scores = [85, 92, 78, 65, 45]

ans = list( filter(lambda x: x>80  , scores))

print(ans)
```

```python
scores = [85, 92, 78, 65, 45]

# def fun1(x):
#   if x>80:
#     return x;

ans = list( map(lambda x: x>80  , scores))

print(ans)
```

# Reduce

reduce(function, iterable)

```python
from functools import reduce

scores = [1, 2, 3, 4, 5]

reduce(lambda x,y: x+y, scores)

```

```python
from functools import reduce

scores = [1, 2, 3, 4, 5]

reduce(lambda x,y: x*y, scores)
```

# Using Context Manger = With

```python
with open('sample2.txt', 'r') as f :
  print(f.read())

print(f.closed)
```

```python
with open('sample2.txt', 'r') as f :
  print(f.read(5))
  print(f.read(5))
```
