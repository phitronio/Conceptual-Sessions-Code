# List + LOOP

### given a list of numbers, find the sum of even numbers and count how many even numbers are there.

```python
numbers = input().split()

even_sum = 0
even_count = 0

for num in numbers:
    num = int(num)
    if num % 2 == 0:
        even_sum += num
        even_count += 1

print("Sum of even numbers:", even_sum)
print("Count of even numbers:", even_count)
```

```python
numbers = list(map(int, input().split()))

even_sum = 0
even_count = 0

for num in numbers:
    if num % 2 == 0:
        even_sum += num
        even_count += 1

print("Sum of even numbers:", even_sum)
print("Count of even numbers:", even_count)
```

# Dictionary

## count frequency of each element in a list using a dictionary.

```python
numbers = input().split()

freq = {}

for num in numbers:
    num = int(num)

    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

# Argument

```python
# 1. Positional Arguments
# arguments are passed based on position (order matters)
def add(a, b):   # a, b are parameters
    print(a + b)

add(10, 20)      # 10 -> a, 20 -> b


# 2. Default Arguments
# parameter has a default value if no argument is given
def greet(name="Guest"):   # default argument
    print("Hello", name)

greet()            # uses default value "Guest"
greet("Akash")     # overrides default value


# 3. Keyword Arguments
# arguments are passed using parameter names (order does not matter)
def student(name, age):
    print(name, age)

student(age=22, name="Rahim")   # explicitly assigning values


# 4. Arbitrary Positional Arguments (*args)
# allows passing multiple values (stored as tuple)
def total_sum(*args):   # args is a tuple
    s = 0
    for i in args:
        s += i
    print(s)

total_sum(10, 20, 30, 40)   # multiple values passed


# 5. Arbitrary Keyword Arguments (**kwargs)
# allows passing multiple key=value pairs (stored as dictionary)

def print_info(**kwargs):   # kwargs is a dictionary  #in ml logging model parameters
    for key, value in kwargs.items():
        print(key, ":", value)

print_info(name="Rahim", age=22, city="Dhaka")
```

# Iterator and Generator

```python
import numpy as np
import pandas as pd

# =========================
# ITERATOR (Easy Example)
# =========================

# Create a numpy array (like dataset features)
arr = np.array([10, 20, 30, 40, 50])

# Convert array to iterator
arr_iter = iter(arr)

# Using next() to manually get values
print(next(arr_iter))  # 10
print(next(arr_iter))  # 20

# In ML / Pandas:
# iterator helps to process data step-by-step instead of all at once






```

```python
# Example with pandas DataFrame
df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

# iterating row by row
for row in df.itertuples():
    print(row)   # each row is accessed one by one
```

```python
# Generator is memory efficient (very important for large dataset in ML)

def data_loader(batch_size, data):
    # simulate large numpy dataset
    for i in range(0, len(data), batch_size):
        yield data[i:i+batch_size]   # yield returns chunk, not whole data

# Large dataset (numpy)
data = np.arange(20)   # [0,1,2,...,19]

loader = data_loader(5, data)

# get batches one by one
print(next(loader))  # first 5 elements
print(next(loader))  # next 5 elements

# ML insight:
# Instead of loading full dataset in memory,
# generator loads data in batches → used in training loop

# Example idea (conceptual):
# for batch in data_loader(32, large_dataset):
#     model.train(batch)
```

# Filter Function

```python
import numpy as np
import pandas as pd

# Create a numpy array (like dataset column)
data = np.array([10, 15, 20, 25, 30])

# Using Python filter with lambda to get values > 20
filtered_np = list(filter(lambda x: x > 20, data))
print("Filtered numpy values:", filtered_np)


# Create a pandas DataFrame
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Score": [85, 90, 78, 92]
})

# Using filter in pandas to get columns starting with 'S'
filtered_columns = df.filter(regex="^S")  # regex for columns starting with S
print("\nFiltered pandas columns:\n", filtered_columns)

# Using boolean filtering in pandas (common in ML)
high_scores = df[df['Score'] > 80]
print("\nRows with Score > 80:\n", high_scores)
```
