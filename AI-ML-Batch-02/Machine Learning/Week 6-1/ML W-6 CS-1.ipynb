```python
# Import Libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
centers = [
    (-5, -5),
    (5, 5),
    (0, 7),
    (6, -4)
]

cluster_std = [1, 1, 1, 1]

X, y = make_blobs(
    n_samples=300,
    centers=centers,
    cluster_std=cluster_std,
    n_features=2,
    random_state=42
)
```

```python
X, y
```

```python
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1])
plt.title("Original 2D Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
km = KMeans(n_clusters=4)
km.fit(X)
```

```python
labels = km.predict(X)
```

```python
labels
```

```python
colors = ["red", "blue", "green", "orange"]

plt.figure(figsize=(7,7))

for i in range(4):
    plt.scatter(
        X[labels == i, 0],
        X[labels == i, 1],
        color=colors[i],
        label=f"Cluster {i+1}"
    )
```

```python
wcss=[]

for k in range(1, 11):
  km = KMeans(n_clusters=k)
  km.fit(X)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(6,5))
plt.plot(range(1,11), wcss, marker="o")
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

# Generate 3D Dataset

```python
centroids = [
    (-5, -5, 5),
    (5, 5, -5),
    (3.5, -2.5, 4),
    (-2.5, 2.5, -4)
]

cluster_std = [1, 1, 1, 1]

X, y = make_blobs(
    n_samples=200,
    centers=centroids,
    cluster_std=cluster_std,
    n_features=3,
    random_state=1
)
```

```python
import plotly.express as px
import pandas as pd
```

```python
fig = px.scatter_3d(
    x=X[:, 0],
    y=X[:, 1],
    z=X[:, 2],
    title="Original 3D Dataset"
)

fig.show()
```

```python
wcss=[]

for k in range(1, 20):
  km = KMeans(n_clusters=k)
  km.fit(X)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(6,5))
plt.plot(range(1,20), wcss, marker="o")
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
km = KMeans(n_clusters=4)
km.fit(X)
```

```python
labels = km.predict(X)
```

```python
result = pd.DataFrame({
    "X": X[:, 0],
    "Y": X[:, 1],
    "Z": X[:, 2],
    "Cluster": labels
})
```

```python
fig = px.scatter_3d(
    result,
    x="X",
    y="Y",
    z="Z",
    color="Cluster",
    title="K-Means Clustering (3D)"
)

fig.show()
```
