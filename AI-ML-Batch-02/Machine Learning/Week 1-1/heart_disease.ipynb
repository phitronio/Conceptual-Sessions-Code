**Step 15 — Outlier Detection using IQR**

```python
q1= df['chol'].quantile(0.25)
q3= df['chol'].quantile(0.75)
print(q1)
print(q3)
iqr= q3-q1
lower_bound= q1-1.5*iqr
upper_bound= q3+1.5*iqr

outliers= df[(df['chol']<lower_bound) | (df['chol']>upper_bound)]
print(outliers)
```
